# Econometric UNFCCC Green Cities Commitment Analysis: USA
## Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('FullDataset.csv')
```

#### Distribution Plots

```python
counts = df['hasCommitments'].value_counts()
plt.figure(figsize = (10,5))
plt.bar(counts.index.astype(str), counts.values, color = sns.color_palette('hls', 2))
plt.title('Distribution of hasCommitments')
plt.xlabel('hasCommitments')
plt.ylabel('Count')
plt.show()
```

```python
counts = df['hasActionsUndertaken'].value_counts()
plt.figure(figsize = (10,5))
plt.bar(counts.index.astype(str), counts.values, color = sns.color_palette('hls', 2))
plt.title('Distribution of hasActionsUndertaken')
plt.xlabel('hasActionsUndertaken')
plt.ylabel('Count')
plt.show()
```

#### Choropleth Plots

```python
plt.figure(figsize = (15,10))
plt.ylim((23,50))
plt.xlim((-127,-65))
cmap = sns.color_palette('vlag', as_cmap = True)
plt.scatter(df['longitude'], df['latitude'], c = df['redState'], cmap = cmap)
plt.title('Choropleth by 2020 Election Results')
plt.xticks([], [])
plt.yticks([], [])
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize = (15,10))
plt.ylim((23,50))
plt.xlim((-127,-65))
cmap = sns.color_palette('Paired', as_cmap = True)
plt.scatter(df.sort_values(by = 'hasCommitments', ascending = True)['longitude'], df.sort_values(by = 'hasCommitments', ascending = True)['latitude'], c = df.sort_values(by = 'hasCommitments', ascending = True)['hasCommitments'], cmap = cmap)
plt.title('Choropleth by Climate Commitments')
plt.xticks([], [])
plt.yticks([], [])
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize = (15,10))
plt.ylim((23,50))
plt.xlim((-127,-65))
cmap = sns.color_palette('Paired', as_cmap = True)
plt.scatter(df.sort_values(by = 'hasActionsUndertaken', ascending = True)['longitude'], df.sort_values(by = 'hasActionsUndertaken', ascending = True)['latitude'], c = df.sort_values(by = 'hasActionsUndertaken', ascending = True)['hasCommitments'], cmap = cmap)
plt.title('Choropleth by Climate Actions Undertaken')
plt.xticks([], [])
plt.yticks([], [])
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize = (18,10))
plt.ylim((23,50))
plt.xlim((-127,-65))
cmap = sns.color_palette('Spectral_r', as_cmap = True)
plt.scatter(df['longitude'], df['latitude'], c = df['povertyProp'], cmap = cmap)
plt.colorbar(shrink = 0.5, pad = 0.025)
plt.title('Choropleth by Proportion of Population in Poverty')
plt.xticks([], [])
plt.yticks([], [])
plt.xlabel('')
plt.ylabel('')
plt.show()
```
